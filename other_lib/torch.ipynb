```python
import numpy as np
import torch
import time
import jax.numpy as jnp
```

```python
def bench(name, func):
    func() 
    start = time.time()
    func()
    end = time.time()
    print(f"{name}: {end - start:.6f} sec")
```

```python
SIZE = 10000  # must match your C++ run
np.random.seed(0)
A_np = np.random.rand(SIZE, SIZE).astype(np.float32)
B_np = np.random.rand(SIZE, SIZE).astype(np.float32)
A_torch = torch.tensor(A_np)
B_torch = torch.tensor(B_np)
```

```python
bench("NumPy matmul", lambda: A_np @ B_np)
bench("PyTorch CPU matmul", lambda: A_torch @ B_torch)
bench("Jax", lambda: jnp.dot(A_np, B_np).block_until_ready())
```
